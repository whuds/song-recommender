```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
features = pd.read_csv('output.csv')
```

```python
features.info()
```

```python
features.drop_duplicates(inplace=True)
```

```python
features.describe()
```

## Important Features
* hotttness
* familiarity
* duration
* loudness
* tempo
* key
* mode
* time_signature

```python
features = features[['hotttness', 'familiarity', 'duration', 'loudness', 'tempo', 'key', 'mode', 'time_signature']]
```

```python
features.describe()
```

```python
features.dropna(inplace=True)
```

```python
features.describe()
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaled = scaler.fit_transform(features)
```

```python
scaled_features = pd.DataFrame(data=scaled,index=features.index, columns=features.columns)
```

```python
scaled_features.describe()
```

```python
import seaborn as sns
plt.figure(figsize=(7,7))
sns.heatmap(scaled_features.corr())
```

```python
sns.pairplot(scaled_features)
```

## Dimensionality Reduction

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, svd_solver='full')
reduced = pca.fit_transform(scaled_features)
```

```python
reduced
```

```python
reduced_df = pd.DataFrame(reduced, columns=['PC1', 'PC2'])
```

```python
sns.jointplot(x='PC1', y='PC2', data=reduced_df)
```

```python
sns.jointplot(x='PC1', y='PC2', data=reduced_df, kind='hex', color='k')
```

```python
from sklearn.cluster import KMeans, SpectralClustering
kmeans = KMeans(n_clusters=3)
clustered = kmeans.fit_predict(reduced_df)
```

```python
clustered
```

```python
LABEL_COLOR_MAP = {0 : 'r',1 : 'g',2 : 'b', 3:'y'}
label_color = [LABEL_COLOR_MAP[l] for l in clustered]
```

```python
plt.figure(figsize = (7,7))
plt.scatter(reduced_df['PC1'], reduced_df['PC2'], c= label_color)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clustering')
plt.show()
```

```python
spectral = SpectralClustering(n_clusters=4, assign_labels='discretize', random_state=0)
clustered_sc = spectral.fit_predict(reduced_df)
LABEL_COLOR_MAP = {0 : 'r',1 : 'g',2 : 'b', 3:'y'}
label_color = [LABEL_COLOR_MAP[l] for l in clustered_sc]
```

```python
plt.figure(figsize = (7,7))
plt.scatter(reduced_df['PC1'], reduced_df['PC2'], c= label_color)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Spectral Clustering: 4 clusters')
plt.show()
```
